```python
import pandas as pd 
import hvplot.pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,1)
y =  x**2 + np.random.randn(10)*2
```

```python
nfit = [val for val in range(1,6)]
xfit = np.arange(0,10)
nfitVals = []
stdestVals = []
colors = ['blue','black','chocolate','purple','olive','lime']
for n in nfit:
    #plt.figure()
    #global stdestVals
    mymodel = np.poly1d(np.polyfit(x,y,n))
    plt.plot(x,y,'go')
    plt.plot(xfit,mymodel(xfit),color = colors[n],label = 'PolyDegree n =  '+str(n))
    plt.legend()
    #residuals
    resid = np.array(y - mymodel(xfit))
    residSquared = resid**2 
    #standard error of estimate SEE
    stdest=np.sqrt(np.sum(residSquared)/(len(residSquared-2)))
     #one dimensional array holding Standard error of estimate values
    stdestVals.append(stdest)
    #one dimensional array holding polynomial fitting degree
    nfitVals.append(n)
    print(mymodel)
    plt.grid(ls = '--',color = 'green')
```

```python
df = pd.DataFrame({'nfit':nfitVals,'stdestVals':stdestVals})
StdEst = df.hvplot.bar(x = 'nfit',y = 'stdestVals',bar_width =0.3,
              grid=True,color='#114422',xlabel='FittingFactor n',          
              ylabel='SEE',rot=0,height =300,width = 550,
              title='StandardErrorOfEstimate')
StdEst
```

```python
print('at index:',np.argmin(stdestVals),'---> minValue of SEE:',
      round(np.min(stdestVals),2))
```
