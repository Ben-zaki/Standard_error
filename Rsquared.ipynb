```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import* 
```

```python
x_data = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y_data = [100,90,80,70,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

### quick  plot to guess polynomial degree for a start

```python
plt.scatter(x_data,y_data,marker='d',s=80,c='purple',label='measured_data')
plt.grid()
plt.legend();
```

```python
df=pd.DataFrame({'time':x_data,'speed':y_data})
```

## select polynomial degree from box

```python
n=BoundedIntText(description='Fit_Degree n',min=1,max=10,step=1)
n
```

```python
mymodel = np.poly1d(np.polyfit(df.time, df.speed, n.value))
speed_pred=mymodel(x_data)
print('fitting_model:')
print(mymodel)
print(65*'*')
```

#### r_squared using numpy 

```python
# predicted - observed
resid_up=speed_pred-np.array(y_data)
```

```python
resid_up_sq=resid_up**2
```

```python
sum_up=resid_up_sq.sum()
```

```python
# mean of observed - observed
resid_down=((np.array(y_data).mean()-np.array(y_data)))
```

```python
resid_down_sq=resid_down**2
```

```python
sum_down=resid_down_sq.sum()
```

```python
r_squared=(1-(sum_up/sum_down))*100
```

```python
print('r_squared :',round(r_squared,2),'%',', degree of poly_fit :',n.value) 
```

#### r_squared using pandas

```python
df_final=pd.DataFrame({'time':x_data,'speed':y_data,'speed_pred':speed_pred,
                 'resid_up':resid_up,'resid_up_sq':resid_up_sq,
                 'resid_down':resid_down,'resid_down_sq':resid_down_sq})
```

```python
r_squared=(1-((resid_up_sq.sum())/(resid_down_sq.sum())))*100
```

```python
plt.figure(figsize=(10,4))
plt.scatter(x_data,y_data,marker='o',s=60,label='measured_data')
plt.scatter(x_data,speed_pred,marker='d',s=100,label='fitted_data')
plt.legend(loc='lower right')
plt.xticks(np.arange(1,23,1))
plt.grid(ls='--',color='gray')
```

#### standard error of estimate

```python
stdest=np.sqrt(resid_up_sq.sum()/(resid_up_sq.size-2))
```

```python
round(stdest,2)
```

#### R-Squared calculation Second method:Sum of (estimated-mean of oberved)^2 / Sum of (observed-mean of oberved)^2

```python
# Predicted - mean of observed
PredictResid = ((np.array(speed_pred)-np.array(y_data).mean())**2).sum()
```

```python
# Observed - mean of 0bserved
ObervedResid = ((np.array(y_data)-np.array(y_data).mean())**2).sum()
```

```python
Rsquared = PredictResid / ObervedResid
Rsquared
print('Polynomial degree : ',n.value)
print('Goodness of fit Rsquared :',round(r_squared,2),'%')
```
